# IMPORTING LIBRARIES

```python
import warnings
warnings.filterwarnings('ignore')

#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing the excel file 

```python
data  = pd.read_excel('NASARAWA TSC.xlsx')
```

```python
plt.figure(figsize = (8,5))

sns.countplot(data['SEX'])

plt.title('Gender Value Counts')
```

```python
data['SEX'].value_counts().plot(kind='pie')
plt.title('Gender Value Counts(Pie Chart)', size = 25, color = 'blue')
plt.figure(figsize = (20,10))
```

#### Female are more employed than male

```python
total = len(data)
plt.figure(figsize = (10, 6))
plt.title('Gender')
ax = sns.countplot(x = 'SEX', data = data)
for p in ax.patches:
    percentage = '{0:.0f}%'.format(p.get_height() / total * 100)
    x = p.get_x() + p.get_width() / 2
    y = p.get_height() + 5
    ax.annotate(percentage, (x, y), ha = 'center')
plt.show()
```

#### The female employment rate is 54%  while the male is 46% 

```python
plt.figure(figsize = (8,5))

sns.countplot(data['LGA'])

plt.title('EMPLOYMENT PER LGA')
```

#### Visualization of the employment per LGA in Nasarawa State

```python
total = len(data)
plt.figure(figsize = (18, 6))
plt.title('EMPLOYMENT PER LGA')
ax = sns.countplot(x = 'LGA', data = data)
for p in ax.patches:
    percentage = '{0:.0f}%'.format(p.get_height() / total * 100)
    x = p.get_x() + p.get_width() / 2
    y = p.get_height() + 5
    ax.annotate(percentage, (x, y), ha = 'center')
plt.show()
```

#### The above visualization shows the percentage of employment per LGA in Nasarawa State were Lafia LGA has the higest percentage followed by Akwanga LGA and others.
